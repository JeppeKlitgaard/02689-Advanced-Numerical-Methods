```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

sns.set_theme(style="whitegrid")

N = 6

fig, axs = plt.subplots(N, figsize=(10, 20))

fig2, ax = plt.subplots(1, figsize=(8, 3))

xjs = 2 * np.pi * np.arange(N) / N
x = np.linspace(0, 2 * np.pi, 250)

for j in range(N):
    xj = xjs[j]

    hjs = (
        1
        / N
        * np.sin(N / 2 * (x - xj))
        * np.cos(1 / 2 * (x - xj))
        / np.sin(1 / 2 * (x - xj))
    )

    axs[j].plot(x, hjs, label=f"h{j}")

    for i in range(N):
        axs[j].axvline(x=xjs[i], color="r", ls="--", lw=1)

        if j == 0:
            ax.axvline(x=xjs[i], color="r", ls="--", lw=1)

    axs[j].set_title(f"h{j}")

    ax.plot(x, hjs, label=f"$h_{j}$")


ax.legend(loc="upper right")
ax.set_xlabel("$x$")
ax.set_ylabel("$h_j(x)$")
ax.set_title(f"Lagrange Polynomials for $N={N}$")
fig2.show()
fig2.savefig("report/output/1c_lagrange_polynomials.png", dpi=600)
```
